```python
import re

mx_to_eu = {
    1.5: 33, 2: 33.5, 2.5: 34, 3: 34.5, 3.5: 35, 4: 36, 4.5: 37,
    5: 37.5, 5.5: 38, 6: 38.5, 6.5: 39, 7: 40, 7.5: 40.5, 8: 41,
    8.5: 42, 9: 42.5, 9.5: 43, 10: 44, 10.5: 44.5, 11: 45, 11.5: 45.5,
    12: 46, 12.5: 47, 13: 47.5, 13.5: 48, 14: 48.5,
    22.5: 36, 23: 37, 23.5: 37.5, 24: 38, 24.5: 38.5, 25: 39,
    25.5: 39.5, 26: 40
}

us_to_eu = {
    3: 35, 3.5: 35.5, 4: 36, 4.5: 37, 5: 37.5, 5.5: 38,
    6: 38.5, 6.5: 39, 7: 40, 7.5: 40.5, 8: 41, 8.5: 42,
    9: 42.5, 9.5: 43, 10: 44, 10.5: 44.5, 11: 45, 11.5: 45.5,
    12: 46, 12.5: 47, 13: 47.5, 13.5: 48, 14: 48.5
}

kids_k_to_eu = {
    1: 16, 1.5: 17, 2: 18, 2.5: 19, 3: 19.5, 3.5: 20, 4: 20.5,
    4.5: 21, 5: 21.5, 5.5: 22, 6: 22.5, 6.5: 23, 7: 23.5,
    7.5: 24, 8: 25, 8.5: 25.5, 9: 26, 9.5: 27, 10: 27.5,
    10.5: 28, 11: 28.5, 11.5: 29, 12: 30, 12.5: 30.5, 13: 31
}

alpha_mapping = {
    'PP': 'XS', 'P': 'S', 'M': 'M', 'G': 'L', 'GG': 'XL',
    '2GG': '2XL', '3GG': '3XL', '4GG': '4XL',
}

western_alpha = {'2XS', 'XS', 'S', 'M', 'L', 'XL', '2XL', '3XL', '4XL', '2X', '3X', '4X', '5XL', '6XL'}

one_size_patterns = {p.upper() for p in [
    'OSFW', 'OSFM', 'OSFL', 'NS', '1', '1 TAILLE', '1 SIZE',
    '1 GRÖßE', '1 GROSSE', 'TAMANHO ÚNICO', 'TAMANHO UNICO',
    'TALLA ÚNICA', 'TALLA UNICA', 'ONE SIZE', 'OS', 'フリー'
]}

def convert_label_string(label_string):
    if not isinstance(label_string, str): return label_string

    individual_sizes = [s.strip() for s in label_string.split('|')]
    eur_sizes = []

    for s in individual_sizes:
        if not s:
            continue

        # 1. One size variants
        if s.upper() in one_size_patterns:
            eur_sizes.append('One Size')
            continue

        # 2. Spanish kids ages with AÑOS (e.g., 7-8 AÑOS, 9-10 años)
        if 'AÑO' in s.upper():
            cleaned = re.sub(r'[Aa][ÑñNn][Oo][Ss]?', 'Y', s).strip()
            eur_sizes.append(cleaned)
            continue

        # 3. Kids age with A suffix (e.g., 7-8A, 9-10A)
        age_a_match = re.match(r'^(\d+-\d+)A$', s)
        if age_a_match:
            eur_sizes.append(f"{age_a_match.group(1)} Y")
            continue

        # 4. CM shoe sizes (e.g., 22.0cm, 28.5cm)
        cm_match = re.match(r'^(\d{2,3}(?:\.\d)?)cm$', s, re.IGNORECASE)
        if cm_match:
            eur_sizes.append(f"EU {cm_match.group(1)}cm")
            continue

        # 5. CM range sizes (e.g., 19-21cm, 22-24cm)
        cm_range_match = re.match(r'^(\d{2}-\d{2})cm$', s, re.IGNORECASE)
        if cm_range_match:
            eur_sizes.append(f"{cm_range_match.group(1)}cm")
            continue

        # 6. Alpha + CM length (e.g., XS 13 CM, S - 8cm, XS - 13cm)
        alpha_cm_match = re.match(r'^(\w+)\s*[-–]?\s*(\d+)\s*cm$', s, re.IGNORECASE)
        if alpha_cm_match:
            base = alpha_cm_match.group(1)
            if base in western_alpha:
                eur_sizes.append(base)
                continue

        # 7. MM shoe sizes (e.g., 220, 225) — Japanese format
        mm_match = re.match(r'^(\d{3})$', s)
        if mm_match:
            val = int(mm_match.group(1))
            if 180 <= val <= 350:
                eur_sizes.append(f"EU {val}mm")
                continue
            # Kids height sizes (e.g., 128, 140, 152)
            if 80 <= val <= 200:
                eur_sizes.append(f"{val}cm (Kids)")
                continue

        # 8. Waist+length combined (e.g., 2430, 3134)
        wl_match = re.match(r'^(\d{2})(\d{2})$', s)
        if wl_match:
            waist = int(wl_match.group(1))
            if 20 <= waist <= 50:
                eur_sizes.append(f"W{waist}")
                continue

        # 9. MX kids ranges (e.g., MX 19-20.5 (niños XXL))
        mx_kids_match = re.match(r'^MX\s+[\d.\s]+-[\d.\s]+\(.*?(XS|S|M|L|XL|XXL|2XL)\)', s, re.IGNORECASE)
        if mx_kids_match:
            size = mx_kids_match.group(1).upper().replace('XXL', '2XL')
            eur_sizes.append(size)
            continue

        # 10. MX shoe sizes (e.g., MX 7, MX 22.5)
        mx_match = re.match(r'^MX\s+(\d+(?:\.\d)?)$', s)
        if mx_match:
            num = float(mx_match.group(1))
            if num in mx_to_eu:
                eur_sizes.append(f"EU {mx_to_eu[num]}")
            continue

        # 11. Kids K sizes (e.g., 4K, 5.5K)
        k_match = re.match(r'^(\d+(?:\.\d)?)K$', s)
        if k_match:
            num = float(k_match.group(1))
            if num in kids_k_to_eu:
                eur_sizes.append(f"EU {kids_k_to_eu[num]}")
            continue

        # 12. Kids K size ranges (e.g., 3K - 5K, 10K - 11.5K)
        k_range_match = re.match(r'^(\d+(?:\.\d)?)K\s*-\s*(\d+(?:\.\d)?)K$', s)
        if k_range_match:
            eur_sizes.append(f"Kids {s}")
            continue

        # 13. Japanese/Asian alpha + inseam (e.g., A/S 7", A/M 3")
        asian_inch = re.match(r'^[JA]/(\w+)\s+\d+"$', s)
        if asian_inch:
            base = asian_inch.group(1)
            if base in western_alpha:
                eur_sizes.append(base)
            continue

        # 14. Japanese/Asian alpha (e.g., J/XS, A/L, A/2XS with optional cup)
        j_match = re.match(r'^[JA]/(\w+)(?:\s*\(.*\))?$', s)
        if j_match:
            base = j_match.group(1)
            if base in western_alpha:
                eur_sizes.append(base)
            continue

        # 15. Bra sizes with parentheses (e.g., XS (A-B), 2XS (C-D))
        bra_paren_match = re.match(r'^(\w+)\s*\([A-Z]{1,2}-[A-Z]{1,2}\)$', s)
        if bra_paren_match:
            base = bra_paren_match.group(1)
            if base in western_alpha:
                eur_sizes.append(base)
            continue

        # 16. Bra sizes without parentheses (e.g., 2XS A-B, XS DD-F)
        bra_no_paren = re.match(r'^(\w+)\s+[A-Z]{1,2}-[A-Z]{1,2}$', s)
        if bra_no_paren:
            base = bra_no_paren.group(1)
            if base in western_alpha:
                eur_sizes.append(base)
            continue

        # 17. Inch waist sizes (e.g., 28x34)
        inch_match = re.match(r'^(\d{2})x\d{2}$', s)
        if inch_match:
            eur_sizes.append(f"W{inch_match.group(1)}")
            continue

        # 18. Inch quote sizes (e.g., 26", 46")
        inch_quote_match = re.match(r'^(\d+)"$', s)
        if inch_quote_match:
            eur_sizes.append(f"{inch_quote_match.group(1)}\"")
            continue

        # 19. Alpha + inch inseam (e.g., XS 5", M 3")
        alpha_inch_match = re.match(r'^(\w+)\s+\d+"$', s)
        if alpha_inch_match:
            base = alpha_inch_match.group(1)
            if base in western_alpha:
                eur_sizes.append(base)
            continue

        # 20. Waist with A prefix (e.g., A24, A28)
        a_waist_match = re.match(r'^A(\d{2})$', s)
        if a_waist_match:
            eur_sizes.append(f"W{a_waist_match.group(1)}")
            continue

        # 21. Waist+length C/L format (e.g., C24/L30, C28 / L30)
        cl_match = re.match(r'^C(\d{2})\s*/\s*L\d+$', s)
        if cl_match:
            eur_sizes.append(f"W{cl_match.group(1)}")
            continue

        # 22. Sock/hat numeric ranges (e.g., 31-33, 34-36)
        sock_match = re.match(r'^(\d{2})-(\d{2})$', s)
        if sock_match:
            eur_sizes.append(s)
            continue

        # 23. US shoe size ranges (e.g., 2 - 3.5, 4.5 - 5.5)
        us_range_match = re.match(r'^(\d+(?:\.\d)?)\s*-\s*(\d+(?:\.\d)?)$', s)
        if us_range_match:
            lo, hi = float(us_range_match.group(1)), float(us_range_match.group(2))
            if lo < 20:
                eur_sizes.append(f"EU {us_to_eu.get(lo, lo)}-{us_to_eu.get(hi, hi)}")
                continue

        # 24. Kids age-based (e.g., 3-4 Y, 0-3 M, ANOS, MESES)
        if 'ANOS' in s.upper() or 'MESES' in s.upper():
            eur_sizes.append(s.upper().replace('ANOS', 'Y').replace('MESES', 'M'))
            continue
        if re.match(r'^\d+-?\d*\s*[YM]$', s):
            eur_sizes.append(s)
            continue

        # 25. Slash-separated alpha (e.g., M / L)
        slash_alpha = re.match(r'^(\w+)\s*/\s*(\w+)$', s)
        if slash_alpha:
            for part in [slash_alpha.group(1), slash_alpha.group(2)]:
                if part in western_alpha:
                    eur_sizes.append(part)
            continue

        # 26. Western alpha (XS, S, M, L, XL, 2XL, 5XL, 6XL...)
        if s in western_alpha:
            eur_sizes.append(s)
            continue

        # 27. Brazilian/PT alpha (P, G, GG, tall variants)
        base_size = re.split(r'[/T]', s)[0]
        if base_size in alpha_mapping:
            eur_sizes.append(alpha_mapping[base_size])
            continue

        # 28. Ball/equipment sizes — skip
        if '号' in s:
            continue

        # 29. Numeric shoe sizes
        try:
            num = float(s)
            if 15 <= num <= 33:
                eur_sizes.append(f"EU {int(num)}")
                continue
            if num in us_to_eu:
                eur_sizes.append(f"EU {us_to_eu[num]}")
                continue
            if 34 <= num <= 54:
                eur_sizes.append(f"EU {int(num)}")
                continue
        except ValueError:
            pass

        # 30. Fractional EU sizes (e.g., 40 2/3, 41 1/3)
        frac_match = re.match(r'^(\d+)\s+(\d)/(\d)$', s)
        if frac_match:
            whole = int(frac_match.group(1))
            frac = int(frac_match.group(2)) / int(frac_match.group(3))
            eur_sizes.append(f"EU {round(whole + frac, 1)}")
            continue

    unique_eur = sorted(list(set(eur_sizes)))
    return " | ".join(unique_eur)

df['size_labels_eur'] = df['size_labels'].apply(convert_label_string)

print(df['size_labels_eur'].value_counts().head(20))
print(f"\nEmpty strings: {(df['size_labels_eur'] == '').sum()}")
```

```python
def bin_badges(text):
    if pd.isna(text): return 'None'
    
    t = text.lower()
    
    # 1. Stock Status (Highest Priority)
    if any(word in t for word in ['sold out', 'esgotado', 'ausverkauft', 'rupture']):
        return 'Out of Stock'
    if any(word in t for word in ['coming soon', 'em breve', '即将发售', 'próximamente']):
        return 'Coming Soon'
    
    # 2. Offers & Exclusivity
    if any(word in t for word in ['exclusive', 'access', '쿠폰', 'adiclub']):
        return 'Member Exclusive'
    if any(word in t for word in ['off', 'discount', 'extra', '直降', 'deal', '할인']):
        return 'On Sale / Promo'
    
    # 3. Product Attributes
    if any(word in t for word in ['novo', 'new', '新品', 'neu', 'nouveau']):
        return 'New Arrival'
    if any(word in t for word in ['pack', 'pack']):
        return 'Multi-Pack'
    if any(word in t for word in ['personaliz', 'custom']):
        return 'Customizable'
    
    # 4. Success Metrics
    if any(word in t for word in ['best seller', 'selling fast', 'winner', 'record']):
        return 'High Performance'
        
    return 'Other'

# Apply to dataframe
df['badge_category'] = df['badge_texts'].apply(bin_badges)
```

```python
def classify_size(size_str):
    if not isinstance(size_str, str) or size_str.strip() == '':
        return 'Unknown'
    
    tokens = [s.strip() for s in size_str.split('|')]
    
    has_one_size = any(t == 'One Size' for t in tokens)
    has_kids_age = any(re.search(r'\d+.*Y$', t) or 'Kids' in t for t in tokens)
    has_eu_cm = any(re.match(r'EU \d+\.\d+cm', t) for t in tokens)  # Japanese cm shoe sizes = adult
    has_eu_mm = any('mm' in t for t in tokens)  # Japanese mm shoe sizes = adult
    has_waist = any(t.startswith('W') and t[1:].isdigit() for t in tokens)
    has_alpha = any(t in {'2XS', 'XS', 'S', 'M', 'L', 'XL', '2XL', '3XL', '4XL', '5XL', '6XL', '2X', '3X', '4X'} for t in tokens)
    has_sock_range = any(re.match(r'^\d{2}-\d{2}$', t) for t in tokens)

    # Extract EU numeric shoe sizes only (not cm/mm)
    eu_nums = []
    for t in tokens:
        m = re.match(r'^EU (\d+(?:\.\d+)?)$', t)  # only pure EU numbers, not EU 22.0cm
        if m:
            eu_nums.append(float(m.group(1)))

    has_eu_kids = eu_nums and max(eu_nums) < 34  # only kids if ALL sizes are small
    has_eu_shoe = bool(eu_nums)

    if has_one_size:
        return 'One Size'
    
    if has_kids_age and not has_eu_shoe:
        return 'Kids'
    
    # JP cm sizes are adult shoes (22cm = EU35, 32cm = EU50)
    if has_eu_cm or has_eu_mm:
        return 'Medium'
    
    if has_eu_shoe:
        if has_eu_kids and not has_kids_age:
            # Could be waist/jeans sizes misread — check range
            if max(eu_nums) >= 28 and min(eu_nums) >= 26:
                return 'Medium'  # likely waist sizes
            return 'Kids'
        avg = sum(eu_nums) / len(eu_nums)
        if avg >= 45:
            return 'Large'
        elif avg >= 38:
            return 'Medium'
        else:
            return 'Kids'
    
    if has_alpha:
        alpha_order = ['2XS', 'XS', 'S', 'M', 'L', 'XL', '2XL', '3XL', '4XL', '5XL', '6XL', '2X', '3X', '4X']
        present = [t for t in tokens if t in set(alpha_order)]
        largest = max(present, key=lambda x: alpha_order.index(x)) if present else None
        if largest in {'4XL', '5XL', '6XL', '4X'}:
            return 'Large'
        elif largest in {'2XL', '3XL', 'XL', '2X', '3X'}:
            return 'Medium'
        else:
            return 'Small'
    
    if has_waist:
        waist_nums = [int(t[1:]) for t in tokens if t.startswith('W') and t[1:].isdigit()]
        if waist_nums and max(waist_nums) >= 36:
            return 'Large'
        return 'Medium'
    
    if has_sock_range:
        return 'Medium'
    
    return 'Unknown'

df_clean['size_bin'] = df_clean['size_labels_eur'].apply(classify_size)
print(df_clean['size_bin'].value_counts())
```
