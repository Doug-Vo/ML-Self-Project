```python
import matplotlib.pyplot as plt

import seaborn as sns

plt.figure(figsize= (12, 6))

plt.subplot(1, 2, 1)
sns.histplot(df['Customer demographics'])
plt.title("Gender Distributiion")

plt.subplot(1, 2, 2)
sns.histplot(df["Product type"])
plt.title("Product Distribution")
plt.tight_layout()
plt.plot()
```

```python
df_melted = df.groupby('Product type')[['Revenue generated', 'Costs']].sum()

df_melted.plot(kind= 'bar')

plt.show()
```

```python
df['is_risk'] = (df['Defect rates'] > 2).astype(int)


risk_count = df['is_risk'].value_counts().sort_index()

plt.pie(risk_count, labels = ["No Risk", "At Risk"], autopct= '%1.1f%%')


plt.show()
```

```python
col_to_encode = ['Product type', 'Customer demographics','Shipping carriers',
                 'Supplier name', 'Location', 'Inspection results', 
                 'Transportation modes', 'Routes']


df = pd.get_dummies(df, columns= col_to_encode, dtype = int, drop_first= True)

df = df.drop(columns = 'SKU')

df.head()
```

```python
df.columns
```

```python
import numpy as np


col_to_corr_1 =['Shipping times', 
              'Availability', 'Price', 'Number of products sold', 
              'Supplier name_Supplier 2',
       'Supplier name_Supplier 3', 'Supplier name_Supplier 4',
       'Supplier name_Supplier 5',
              'Stock levels', 'Production volumes', 'Manufacturing lead time',
              'Costs', 'Transportation modes_Rail',
              'Transportation modes_Road', 'Transportation modes_Sea',
              ]



matrix = df[col_to_corr_1].corr()
mask = np.triu(np.ones_like(matrix, dtype = bool))
sns.heatmap(matrix, mask = mask, cmap = 'coolwarm', 
            annot = True, center = 0, fmt= '.2f', cbar= False)


plt.show()
```

```python
import numpy as np


col_to_corr_2 =['Lead times', 'Availability', 
              'Stock levels', 'Production volumes',
              'Costs', 'Transportation modes_Rail',
              'Transportation modes_Road', 'Transportation modes_Sea',
              'Shipping carriers_Carrier B',
       'Shipping carriers_Carrier C'
              ]




matrix = df[col_to_corr_2].corr()
mask = np.triu(np.ones_like(matrix, dtype = bool))
sns.heatmap(matrix, mask = mask, cmap = 'coolwarm', 
            annot = True, center = 0, fmt= '.2f', cbar= False)


plt.show()
```
