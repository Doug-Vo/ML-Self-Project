```python
import pandas as pd

df = pd.read_csv('Data/supply_chain_data.csv')

df
```

```python
for col in df.columns:
    print(col)
    if df[col].nunique() < 5:
        print(df[col].unique())
    else:
        print(df[col].nunique())

    print(f"Min: {df[col].min()}")
    print(f"Max: {df[col].max()}")
    print("-------------------\n")
```

```python
import matplotlib.pyplot as plt

import seaborn as sns

plt.figure(figsize= (12, 6))

plt.subplot(1, 2, 1)
sns.histplot(df['Customer demographics'])
plt.title("Gender Distributiion")

plt.subplot(1, 2, 2)
sns.histplot(df["Product type"])
plt.title("Product Distribution")
plt.tight_layout()
plt.plot()
```

```python

plt.figure(figsize= (8, 6))
sns.countplot(df, x= 'Customer demographics', hue = 'Product type')

plt.show()
```

```python
df.columns
```

```python
plt.figure(figsize= (12, 6))

plt.subplot(1, 2, 1)
sns.histplot(df['Inspection results'])


plt.subplot(1, 2, 2)
sns.histplot(df["Defect rates"])
plt.tight_layout()
plt.plot()
```

```python
sns.countplot(df, x = 'Routes', hue = 'Transportation modes')
plt.show()
```

```python
df.columns
```

```python
df_melted = df.groupby('Product type')[['Revenue generated', 'Costs']].sum()

df_melted.plot(kind= 'bar')

plt.show()
```

```python
df['is_risk'] = (df['Defect rates'] > 2).astype(int)


risk_count = df['is_risk'].value_counts().sort_index()

plt.pie(risk_count, labels = ["No Risk", "At Risk"], autopct= '%1.1f%%')


plt.show()
```

```python
df.info()
```

```python
col_to_encode = ['Product type', 'Customer demographics','Shipping carriers',
                 'Supplier name', 'Location', 'Inspection results', 
                 'Transportation modes', 'Routes']


df = pd.get_dummies(df, columns= col_to_encode, dtype = int)

df = df.drop(columns = 'SKU')

df.head()
```

```python
import numpy as np

col = df.columns


matrix = df.corr()
mask = np.triu(np.ones_like(matrix, dtype = bool))
sns.heatmap(matrix, mask = mask, cmap = 'coolwarm', 
            annot = False, center = 0)


plt.show()
```
