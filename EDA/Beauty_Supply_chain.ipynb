## Cluster

```python
df.columns
```

```python
from sklearn.preprocessing import StandardScaler


scaler = StandardScaler()


features_to_cluster = [
    'Price', 'Number of products sold',
    'Revenue generated', 
    'Stock levels',
    'Costs'
]

cluster_df  = scaler.fit_transform(df[features_to_cluster])
```

```python
from sklearn.decomposition import PCA


scale = 3
pca = PCA(n_components= 2)

pca_data = pca.fit_transform(cluster_df)


loadings = pca.components_.T * np.sqrt(pca.explained_variance_)

plt.figure(figsize= (12, 8))

plt.scatter(pca_data[:, 0], pca_data[:, 1], alpha = 0.5)

for i, feature in enumerate(features_to_cluster):
    x_val = loadings[i, 0] * scale
    y_val = loadings[i, 1] * scale
    plt.arrow(0, 0, x_val, y_val, color = 'red')

    plt.text(x_val * 1.1, y_val * 1.1, feature, fontsize = 8)


plt.show()
```

Our PCA shows us the distinct features that can help identify the uniqueness of our data

```python
from sklearn.cluster import KMeans
from kneed import KneeLocator



inertia = []

K_range = range(1, 10)


for k in K_range:
    kmeans = KMeans(n_clusters= k, random_state= 8, n_init= 10)
    kmeans.fit(cluster_df)
    inertia.append(kmeans.inertia_)

kl = KneeLocator(K_range, inertia, curve= 'convex', direction='decreasing')
optimal_k = kl.elbow

plt.plot(K_range, inertia, marker = 'o', linestyle = '--')
plt.axvline(optimal_k, color = 'red', linestyle = ':', label = f'Optimal k: {optimal_k} ')
plt.xlabel("Number of Cluster")
plt.ylabel("Intertia")

plt.grid(True)
plt.legend()
plt.show()

print(f"Optimal k: {kl.elbow}")
```

```python
model = KMeans(n_clusters= optimal_k, random_state= 8)

labels = model.fit_predict(cluster_df)

final_df = df.copy()

final_df['Cluster'] = labels
print(final_df.groupby('Cluster')[features_to_cluster].mean())
```



### **Cluster 0: High-Value / High-Cost (The Luxury Tier)**

* **Profile:** High Price ($81.41) & Highest Revenue ($6,803).
* **The Problem:** Highest average **Costs ($721.49)**.


### **Cluster 1: Low-Demand / Low-Price (The Underperformers)**

* **Profile:** Lowest Sales (220 units) & Lowest Price ($24.45).
* **The Problem:** Stagnant inventory with low return.


### **Cluster 2: Overstocked / Capital-Heavy (The Inventory Trap)**

* **Profile:** High Price ($73.08) & Highest **Stock Levels (77.7)**.
* **The Problem:** Massive amount of capital tied up in sitting inventory.


### **Cluster 3: High-Velocity / Lean (The Efficiency Leaders)**

* **Profile:** Highest **Sales (801 units)** & Leanest Stock (27.2).
* **The Problem:** High risk of **Stockouts**.
